**Description:** Downloading the FC25 Dataset into a pandas dataframe and removes unnecessary columns.

```python
# import libraries
import pandas as pd
```

```python
fc25_df = pd.read_csv("/Users/Jake/Documents/GitHub/Data-Wrangling-Project/data/raw/FC25 Dataset/male_players.csv", encoding="utf-8")
display(fc25_df)
```

```python
# Remove unnecessary columns
new_df = fc25_df.loc[:, ["Name", "OVR", "PAC", "SHO", "PAS", "DRI", "DEF", "PHY", "Nation", "League", "Team", "Age", "Height", "Weight", "Position"]]
display(new_df)
```

```python
# Replace notable players that are spelled differently between datasets

# Change vini jr to Vinicius J√∫nior
new_df["Name"] = new_df["Name"].str.replace("Vini Jr.", "Vinicius J√∫nior")

# Change Gabriel to Gabriel Magalh√£es
new_df["Name"] = new_df["Name"].str.replace("Gabriel", "Gabriel Magalh√£es")

# Change Pierre-Emile H√∏jbjerg to Pierre H√∏jbjerg
new_df["Name"] = new_df["Name"].str.replace("Pierre-Emile H√∏jbjerg", "Pierre H√∏jbjerg")

# Change Heung Min Son to Son Heung-min
new_df["Name"] = new_df["Name"].str.replace("Heung Min Son", "Son Heung-min")

# Change Fran Garc√≠a to Fran Garcia
new_df["Name"] = new_df["Name"].str.replace("Fran Garc√≠a", "Fran Garcia")

# Change Julio C√©sar Enciso to Julio Enciso
new_df["Name"] = new_df["Name"].str.replace("Julio C√©sar Enciso", "Julio Enciso")

# Change Brahim to Brahim D√≠az
new_df["Name"] = new_df["Name"].str.replace("Brahim", "SBrahim D√≠az")

# Change Carvajal to Dani Carvajal
new_df["Name"] = new_df["Name"].str.replace("Carvajal", "Dani Carvajal")

# Change Balde to Alejandro Balde
new_df["Name"] = new_df["Name"].str.replace("Balde", "Alejandro Balde")

# Change Ronald Araujo to Ronald Ara√∫jo
new_df["Name"] = new_df["Name"].str.replace("Ronald Araujo", "Ronald Ara√∫jo")

# Change Romeo Lavia to Rom√©o Lavia
new_df["Name"] = new_df["Name"].str.replace("Romeo Lavia", "Rom√©o Lavia")

# Change Oyarzabal to Mikel Oyarzabal
new_df["Name"] = new_df["Name"].str.replace("Oyarzabal", "Mikel Oyarzabal")

# Change Palhinha to Jo√£o Palhinha
new_df["Name"] = new_df["Name"].str.replace("Palhinha", "Jo√£o Palhinha")

# Change Savinho to S√°vio
new_df["Name"] = new_df["Name"].str.replace("Savinho", "S√°vio")

# Change Benjamin White to Ben White
new_df["Name"] = new_df["Name"].str.replace("Benjamin White", "Ben White")

# Change J√©r√©my Doku to Jeremy Doku
new_df["Name"] = new_df["Name"].str.replace("J√©r√©my Doku", "Jeremy Doku")

# Change Zubimendi to Mart√≠n Zubimendi
new_df["Name"] = new_df["Name"].str.replace("Zubimendi", "Mart√≠n Zubimendi")

# Change Giovanni Reyna to Gio Reyna
new_df["Name"] = new_df["Name"].str.replace("Giovanni Reyna", "Gio Reyna")

# Change Ferran Torres to Ferr√°n Torres
new_df["Name"] = new_df["Name"].str.replace("Ferran Torres", "Ferr√°n Torres")
```
