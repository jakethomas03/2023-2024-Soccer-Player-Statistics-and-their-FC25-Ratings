**Description:** Merging the FBref dataset with FC25 dataset.

```python
# import libraries
import pandas as pd
```

```python
fbref_df = pd.read_csv("/Users/Jake/Documents/GitHub/Data-Wrangling-Project/data/working/fbref_data_clean.csv")
fc25_df = pd.read_csv("/Users/Jake/Documents/GitHub/Data-Wrangling-Project/data/working/fc25_data_clean.csv")
display(fbref_df)
display(fc25_df)
```

```python
# Remove unnamed columns
fbref_df = fbref_df.loc[:, ~fbref_df.columns.str.contains('^Unnamed')]
fc25_df = fc25_df.loc[:, ~fc25_df.columns.str.contains('^Unnamed')]
display(fbref_df)
display(fc25_df)
display(fbref_df)
```

```python
# Rename primary key column name
fc25_df.rename(columns={"Name": "Player"}, inplace=True)
```

```python
# Strip whitespaces 
fbref_df["Player"] = fbref_df["Player"].str.strip()
fc25_df["Player"] = fc25_df["Player"].str.strip()
```

```python
merged_df = pd.merge(fbref_df, fc25_df, on="Player", how="inner")
display(merged_df)
```

```python
# Drop duplicates
merged_df.drop_duplicates(subset=["Player"], keep='first', inplace=True)
merged_df.reset_index()

display(merged_df)

```

```python
# Remove unnecessary columns
merged_df = merged_df.loc[:, ["Player", "OVR", "Goals", "Assists", "Expected Goals", "PAC", "SHO", "PAS", "DRI", "DEF", "PHY", "Nation", "League_x", "Club Team", "Age_y", "Height", "Weight", "Position_y"]]
display(merged_df)
```

```python
# Rename columns
merged_df.rename(columns={
    'OVR': 'Overall',
    'PAC': 'Pace',
    'SHO': 'Shooting',
    'PAS': 'Passing',
    'DRI': 'Dribbling',
    'DEF': 'Defending',
    'PHY': 'Physicality',
    'Nation': 'National Team',
    'League_x': 'League',
    'Age_y': 'Age',
    'Position_y': 'Position', 
    'Height': 'Height_cm', 
    'Weight': 'Weight_kg', 
}, inplace=True)
```

```python
# Change league names
merged_df["League"] = merged_df["League"].str.replace("it Serie A", "Serie A")

merged_df["League"] = merged_df["League"].str.replace("eng Premier League", "Premier League")

merged_df["League"] = merged_df["League"].str.replace("fr Ligue 1", "Ligue 1")

merged_df["League"] = merged_df["League"].str.replace("de Bundesliga", "Bundesliga")

merged_df["League"] = merged_df["League"].str.replace("es La Liga", "La Liga")

display(merged_df)
```

```python
# Change height and weight to numeric
merged_df['Height_cm'] = merged_df['Height_cm'].str.split('c').str[0]
merged_df['Weight_kg'] = merged_df['Weight_kg'].str.split('k').str[0]

merged_df['Height_cm'] = merged_df['Height_cm'].astype(int)
merged_df['Weight_kg'] = merged_df['Weight_kg'].astype(int)
display(merged_df)
```

```python
# Save new merged/cleaned dataframe as new csv
merged_df.to_csv("merged_final_dataset.csv", index=False, encoding="utf-8")
```
